```python
import pandas as pd

df = pd.read_csv('311data_202104_present.csv')
pd.set_option("display.max_columns", None)
#adding commas to numbers (ef 123,345)

# Showing ALL columns
pd.set_option ("display.max_columns", None)
df.head()
```

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.info()
```

```python
df.head()
```

```python
df.complaint_type.value_counts().head(60)
```

```python
ro = df[df.complaint_type == "Rodent"]
ro.head()
```

```python
ro.descriptor.value_counts()
```

```python
ratmo = ro[(ro.descriptor == "Rat Sighting") | (ro.descriptor == "Mouse Sighting") ]
ratmo.head()
```

```python
ratmo.borough.value_counts().plot(kind ="barh")
```

```python
ratmo.created_date = pd.to_datetime(ratmo['created_date'])
ratmo.closed_date = pd.to_datetime(ratmo['closed_date'])
```

```python
ratmo.info()
```

```python
ratmo.resample('M', on='created_date').size().plot()
```

```python
ratmo.to_csv('ratmouse.csv')
ratmo.to_csv('ratmouse.csv',index=False)
```
